```python
def calculate_f1_score(precision, recall):
    F1 = 2 * (precision * recall) / (precision + recall)
    return F1
```

#**Про сходство датасетов в первом смысле**

По введенному определению сходства датасетов в первом смысле, модель, обученная на датасете A и показывающая хорошие результаты при тестировании на A, также показывает хорошие результаты при тестировании и на B.

То есть проанализировать нужно графики слева-внизу и справа-вверху.

Анализ графиков:

**1) при обучении на датасете с отзывами о фильмах и тестировании на датасете с отзывами об отелях (переход с отзывов о фильмах на отзывы об отелях)**:

**I. качество на тестовой выборке:**

Значения accuracy при тестировании в зависимости от модели колеблются  примерно от 0.51 до 0.87, среднее арифметическое и медианное значения в районе 0.7.

Значения f1_score при тестировании колеблются примерно от 0.4 до 0.87; среднее арифметическое значение f1_score примерно 0.7, тогда как медианное значение f1_score в районе 0.75. У моделей ALBERT и talkheads_ggelu_bert наблюдаются наименьшие значения f1_score: примерно 0.4 и 0.46 соответственно. Для остальных моделей f1_score не менее 0.65.


**II. абсолютное изменение качества:**

#Дополнительное исследование сходства датасетов во втором смысле

В этой части мы исследуем сходство датасетов во втором смысле подробнее.

Мы изучим, является ли разница в качестве стабильной (для разных моделей) при переходе

 **_от_** обучения и тестирования на датасете А
 
  **_к_** обучению и тестированию на датасете В.

Рассматриваем абсолютное изменение качества для каждой модели при переходе с датасета movie rewiews на датасет hotel reviews:

(_accuracy on hotel reviews after training on hotel reviews_) – (_accuracy on movie reviews after training on movie reviews_)

и

(_f1_score on hotel reviews after training on hotel reviews_) – (_f1_score on movie reviews after training on movie reviews_).


```python
df_with_test_results = df_with_all_models
model_names_lower_case_list = ['albert', 'bert', 'distilbert', 'electra', 'labse', 'lambert', 'mobilebert', 'roberta', 'talkheads_ggelu_bert', 'tn_bert']
```

```python
df_different_datasets_test_quality_absolute_difference = pd.DataFrame(columns=['accuracy', 'f1_score'])
adding = ' hotel reviews vs movie reviews absolute difference on test'
for model_name in model_names_lower_case_list:
    index_name_1 = model_name + ' trained on hotel reviews and tested on hotel reviews dataset'
    index_name_2 = model_name + ' trained on movie reviews and tested on movie reviews dataset'
    df_different_datasets_test_quality_absolute_difference.loc[model_name + adding, 'accuracy'] = df_with_test_results.loc[index_name_1, 'accuracy'] - df_with_test_results.loc[index_name_2, 'accuracy']
    df_different_datasets_test_quality_absolute_difference.loc[model_name + adding, 'f1_score'] = df_with_test_results.loc[index_name_1, 'f1_score'] - df_with_test_results.loc[index_name_2, 'f1_score']
```
