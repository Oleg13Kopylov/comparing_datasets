```python
cur_name_for_dataframe = 'df_with_test_results_for_first_series_of_experiments'
df_with_test_results_for_first_series_of_experiments = pd.read_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + cur_name_for_dataframe + '.csv')
df_with_test_results_for_first_series_of_experiments.set_index('Unnamed: 0', inplace=True) 
df_with_test_results_for_first_series_of_experiments.index.name = None 
df_with_test_results_for_first_series_of_experiments = df_with_test_results_for_first_series_of_experiments[
    df_with_test_results_for_first_series_of_experiments.index.str.contains('on hotel reviews and tested on hotel reviews dataset') |
     df_with_test_results_for_first_series_of_experiments.index.str.contains('on movie reviews and tested on movie reviews dataset') == True
     ]
```

```python
cur_name_for_dataframe = 'df_with_test_results_for_second_series_of_experiments'
df_with_test_results_for_second_series_of_experiments = pd.read_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + cur_name_for_dataframe + '.csv')
df_with_test_results_for_second_series_of_experiments.set_index('Unnamed: 0', inplace=True) 
df_with_test_results_for_second_series_of_experiments.index.name = None 
df_with_test_results_for_second_series_of_experiments = df_with_test_results_for_second_series_of_experiments[
    df_with_test_results_for_second_series_of_experiments.index.str.contains('on spam sms tested on spam sms dataset') |
     df_with_test_results_for_second_series_of_experiments.index.str.contains('on spam emails tested on spam emails dataset') == True
     ]
```

```python
df_with_test_results_for_second_series_of_experiments
```

```python
df_with_test_results_for_spam_sms = df_with_test_results_for_second_series_of_experiments[
    df_with_test_results_for_second_series_of_experiments.index.str.contains('on spam sms tested on spam sms dataset')
     ]
```

```python
df_with_test_results_for_spam_emails = df_with_test_results_for_second_series_of_experiments[
    df_with_test_results_for_second_series_of_experiments.index.str.contains('on spam emails tested on spam emails dataset')
]
```

```python
df_with_test_results_for_hotel_reviews = df_with_test_results_for_first_series_of_experiments[
    df_with_test_results_for_first_series_of_experiments.index.str.contains('on hotel reviews and tested on hotel reviews dataset')
]
```

```python
df_with_test_results_for_movie_reviews = df_with_test_results_for_first_series_of_experiments[
     df_with_test_results_for_first_series_of_experiments.index.str.contains('on movie reviews and tested on movie reviews dataset')
     ]
```

```python
model_names_lower_case_list = ['albert', 'bert', 'distilbert', 'electra', 'labse', 'lambert', 'mobilebert', 'roberta', 'talkheads_ggelu_bert', 'tn_bert']
model_names = ['ALBERT', 'BERT', 'DistilBERT', 'ELECTRA', 'LaBSE',
               'LAMBERT', 'MobileBERT', 'ROBERTA', 'talkheads_ggelu_bert', 'tn_bert']
```

# Часть 1: датасет с отзывами об отелях VS датасет с SMS

Рассматриваем абсолютное изменение качества для каждой модели при переходе с датасета hotel reviews на датасет spam sms:

(f1_score on spam sms after training on spam sms) – (f1_score on hotel reviews after training on hotel reviews).

```python
df_different_datasets_test_quality_absolute_difference = pd.DataFrame(columns=['f1_score'])
# рассматриваем разницу качеств для каждой модели:
# (f1_score on spam sms after training on spam sms) – (f1_score on hotel reviews after training on hotel reviews).
adding = ' hotel reviews vs sms absolute difference on test'
for model_name in model_names_lower_case_list:
    index_name_1 = model_name + ' trained on spam sms tested on spam sms dataset'
    index_name_2 = model_name + ' trained on hotel reviews and tested on hotel reviews dataset'
    first = float(df_with_test_results_for_spam_sms.loc[index_name_1, 'f1_score'])
    second = float(df_with_test_results_for_hotel_reviews.loc[index_name_2, 'f1_score'])
    df_different_datasets_test_quality_absolute_difference.loc[model_name + adding, 'f1_score'] = first - second
df_different_datasets_test_quality_absolute_difference
```

```python
name_for_dataframe = 'df_absolute_quality_difference_when_switching_from_hotel_reviews_to_sms'
df_different_datasets_test_quality_absolute_difference.to_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + name_for_dataframe + '.csv')
```

Сохраним также аналогичный датафрейм, но уже для перехода в другую сторону: с sms на hotel_reviews.

```python
other_direction_df_different_datasets_test_quality_absolute_difference = -df_different_datasets_test_quality_absolute_difference
name_for_dataframe = 'df_absolute_quality_difference_when_switching_from_sms_to_hotel_reviews'
other_direction_df_different_datasets_test_quality_absolute_difference.to_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + name_for_dataframe + '.csv')
```

Визуализируем 1: абсолютное изменение качества для каждой модели при переходе с датасета hotel reviews на датасет spam sms (подробно описано выше)

```python
fig, axes = plt.subplots(2, figsize=(14, 12))

axes[0].scatter(x=model_names, y=df_different_datasets_test_quality_absolute_difference.f1_score, label='f1_score', s=100)
axes[0].tick_params(axis='x', labelrotation = 35)
axes[0].set_title('f1_score absolute difference for each model trained on hotel reviews and tested on hotel reviews VS trained on sms and tested on sms', fontsize=16)
axes[0].legend()

axes[1].scatter(x=['f1_score'] * len(df_different_datasets_test_quality_absolute_difference.f1_score), y=df_different_datasets_test_quality_absolute_difference.f1_score)
axes[1].scatter(x=['f1_score'], y=[df_different_datasets_test_quality_absolute_difference.f1_score.mean()], color='blue', marker='s', s=100, label='Mean value')
axes[1].scatter(x=['f1_score'], y=[df_different_datasets_test_quality_absolute_difference.f1_score.median()], color='black', marker='x', s=150, label='Median value')
axes[1].set_title('f1_score absolute difference for models trained on hotel reviews and tested on hotel reviews VS trained on sms and tested on sms', fontsize=16)
axes[1].legend()

fig.tight_layout()
plt.show()
```

Теперь рассматриваем относительное изменение качества, которое выражено в процентах:

```python
df_different_datasets_test_quality_relative_difference_in_percent = pd.DataFrame(columns=['f1_score'])
other_direction_df_different_datasets_test_quality_relative_difference_in_percent = pd.DataFrame(columns=['f1_score']) # new
adding = ' hotel reviews vs sms relative difference (in percent) on test'
other_direction_adding = ' sms vs hotel reviews relative difference (in percent) on test' # new
for model_name in model_names_lower_case_list:
    index_name_1 = model_name + ' trained on spam sms tested on spam sms dataset'
    index_name_2 = model_name + ' trained on hotel reviews and tested on hotel reviews dataset'
    first = float(df_with_test_results_for_spam_sms.loc[index_name_1, 'f1_score'])
    second = float(df_with_test_results_for_hotel_reviews.loc[index_name_2, 'f1_score'])
    df_different_datasets_test_quality_relative_difference_in_percent.loc[model_name + adding, 'f1_score'] = (first - second) /  second * 100
    # new
    other_direction_df_different_datasets_test_quality_relative_difference_in_percent.loc[model_name + other_direction_adding, 'f1_score'] = (second - first) / first * 100
```

```python
df_different_datasets_test_quality_relative_difference_in_percent
```

```python
name_for_dataframe = 'df_relative_quality_difference_in_percent_when_switching_from_hotel_reviews_to_sms'
df_different_datasets_test_quality_relative_difference_in_percent.to_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + name_for_dataframe + '.csv')
```

```python
# new
name_for_dataframe = 'df_relative_quality_difference_in_percent_when_switching_from_sms_to_hotel_reviews'
other_direction_df_different_datasets_test_quality_relative_difference_in_percent.to_csv('/content/drive/MyDrive/data_for_colab/dataframes/additional_research/' + name_for_dataframe + '.csv')
```

Визуализируем 2: относительное изменение качества для каждой модели при переходе с датасета hotel reviews на датасет spam sms, выраженное в процентах

```python
fig, axes = plt.subplots(2, figsize=(14, 12))

axes[0].scatter(x=model_names, y=df_different_datasets_test_quality_relative_difference_in_percent.f1_score, label='f1_score', s=100)
axes[0].tick_params(axis='x', labelrotation = 35)
axes[0].set_title('f1_score relative difference (in percent) for each model trained on hotel reviews and tested on hotel reviews VS trained on sms and tested on sms', fontsize=16)

axes[1].scatter(x=['f1_score'] * len(df_different_datasets_test_quality_relative_difference_in_percent.f1_score), y=df_different_datasets_test_quality_relative_difference_in_percent.f1_score)
axes[1].scatter(x=['f1_score'], y=[df_different_datasets_test_quality_relative_difference_in_percent.f1_score.mean()], color='blue', marker='s', s=100, label='Mean value')
axes[1].scatter(x=['f1_score'], y=[df_different_datasets_test_quality_relative_difference_in_percent.f1_score.median()], color='black', marker='x', s=150, label='Median value')
axes[1].set_title('f1_score relative difference (in percent) for models trained on hotel reviews and tested on hotel reviews VS trained on sms and tested on sms', fontsize=16)

plt.legend()
fig.tight_layout()
plt.show()
```

**Анализ:**

1) Абсолютное изменение f1_score при переходе с датасета hotel reviews на датасет spam sms составляет от 0.02 до 0.22 (в зависимости от модели). Среднее значение и медианное значение приблизительно 0.1075.
 
2) Относительное изменение f1_score при переходе с датасета hotel reviews на датасет spam sms составляет от 2% до 31%. Среднее значение около 15%, тогда как медианное значение приблизительно 14%.

------------------
Таким образом, **абсолютное и относительное изменение качества при переходе с датасета hotel reviews на датасет spam sms достаточно сильно отличается от модели к модели. Сходство датасетов во втором смысле не наблюдается.**
